2. Explain the difference between Absolute and Relative positioning.

  Relative Positioning (position: relative;)

  An element with position: relative; is positioned relative to its original position in the normal document flow.
  If you don’t set any positioning offsets (like top, left, right, or bottom), the element appears in its usual spot.
  If you do set offsets, the element will move relative to where it would have been in the normal flow. For example, top: 10px will shift it 10 pixels down from its original position.
  The space it originally occupied in the page layout is still preserved, even if the element is moved visually.

  Absolute Positioning (position: absolute;)

  An element with position: absolute; is taken out of the normal document flow, meaning it no longer affects (and is not affected by) other elements’ positioning.
  The element is positioned relative to its first ancestor that has a positioning context (i.e., an ancestor that is positioned relative, absolute, or fixed). If no such ancestor is found, it’s positioned relative to the browser window (the <body>).
  Offsets like top: 10px; left: 20px; place the element exactly 10px from the top and 20px from the left edge of the positioned ancestor.
  Because it’s out of the normal flow, surrounding elements won’t shift to accommodate it, and it can overlap other elements.

5. Explain the z-index, with a code example.

  The z-index property in CSS controls the stacking order of elements that overlap. An element with a higher z-index is stacked on top of elements with lower z-index values.

  Positioning: z-index only works on elements that have a position value other than static (i.e., position: relative;, position: absolute;, position: fixed;, or position: sticky;).

  Stacking Context: Each positioned element (or a few other special cases like opacity < 1) can create a new stacking context. Within that context, child elements’ z-index values are compared with each other, not with elements outside that context.

```
<!DOCTYPE html>
<html>
<head>
  <meta charset="UTF-8" />
  <title>Z-Index Example</title>
  <style>
    .box {
      width: 100px;
      height: 100px;
      position: absolute; /* Important for z-index to work */
    }
    .box1 {
      background-color: red;
      top: 50px;
      left: 50px;
      z-index: 1; /* Lower stacking order */
    }
    .box2 {
      background-color: green;
      top: 80px;
      left: 80px;
      z-index: 2; /* Higher stacking order, sits on top of .box1 */
    }
    .box3 {
      background-color: blue;
      top: 110px;
      left: 110px;
      z-index: 3; /* Even higher stacking order */
    }
  </style>
</head>
<body>

  <div class="box box1">Box 1</div>
  <div class="box box2">Box 2</div>
  <div class="box box3">Box 3</div>

</body>
</html>
```
